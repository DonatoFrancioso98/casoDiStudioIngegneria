```python
#Classe che gestisce gli elementi di una coda
class codaE:
    def __init__(self,g, f):
        self.g = g
        self.f = f
    def getG(self):
        return self.g
    def getF(self):
        return self.f
    def setG(self,g):
        self.g = g
    def setF(self,f):
        self.f = f

#Classe coda prioritaria per la gestione della lista di open nell'algoritmo A*     
#Gli elementi sono messi in ordine crescente in base al loro peso dall'entrata al nodo corrente
class pQueue(codaE):
    def __init__(self):
        self.queue = []
  
    def __str__(self):
        return ' '.join([str(i) for i in self.queue])
  
    #restituisce true se la coda è vuota
    def isEmpty(self):
        return len(self.queue) == 0
    
    #inserimento di un elemento all'interno della coda
    def insert(self, data):
        self.queue.append(data)
    
    #restituisce 1 se l'elemento passato per parametro è presente nella coda
    def presente(self, str):
        for i in range(len(self.queue)):
            if str == self.queue[i].getG().getNome():
                return 1
            
        return 0
    
    #stampa della coda
    def stampa(self):
        for i in range(len(self.queue)):
            tmp = self.queue[i].getG().getNome()
            print(tmp)

    #elimina e restituisce il valore più piccolo della coda        
    def delete(self):
        try:
            primo = 0
            for i in range(len(self.queue)):
                g1 = self.queue[i].getF()
                if primo == 0:
                    min = i
                    primo = 1
                else:
                    if g1 < self.queue[min].getF():
                        min = i
            item = self.queue[min]
            del self.queue[min]
            return item
        except IndexError:
            print()
            exit()
```

```python
#Classe per la gestione dell'algoritmo A* della ricerca del cammino minimo
#L'algoritmo restituisce il percorso più breve da svolgere all'interno di una sezione restituita dai dati inseriti da input dallìutente
#Il cammimo pià breve viene calcolato sia in base al peso degli archi che al tempo di attesa della giostra
class Astar:
    def __init__(self, start, goal):
        self.start = start
        self.goal = goal
        


    def Pitagora(self, g1, g2):
        x1 = g1.punto.getX()
        y1 = g1.punto.getY()
        x2 = g2.punto.getX()
        y2 = g2.punto.getY()
        pit = math.sqrt(((x2 - x1)**2) + ((y2 - y1)**2))
        return pit

    #20 sono i metri massimi ogni quanto è presente un'attrazione
    def HeuristicCost(self, v):
        if(v.getNome() == "Uscita"):
            return 0
        else:
            pit = self.Pitagora(v, self.goal)
            x = pit * 10
            y = int(x/20)
            if y == 0:
                tempo = 0
            else:
                tempo = y * 11 
            return x + tempo
         
    def restituisciFloat(self, g):
        for i in self.pippo:
            if i == g.getName():
                return i
    
    def presInClose(self, str, coda):
        for i in range(len(coda)):
            if str == coda[i].getG().getNome():
                return 1
            
        return 0

    def svolgimento(self):
        tentative = 0
        came_from: dict[str][str] = {}
        minimo = sys.maxsize
        minimoVecchio = sys.maxsize
        close = []
        open = pQueue()
        tmp = codaE(self.start, 0)
        open.insert(tmp)
        while open.isEmpty() == 0:
            current = open.delete()
            if current.getG().getNome() == "Uscita":
                break
            tentative = 0
            for next in listaVicini(current.getG()):
                tentative = 0
                new_cost = current.getG().getCost() + next.getTas() + getPeso(current.getG(), next)
                if open.presente(next.getNome()) == 0 and self.presInClose(next.getNome(), close) == 0: #and if next not in open:
                    tentative = 1
                    next.setCost(new_cost)
                else:
                    if new_cost < next.getCost():
                        next.setCost(new_cost)
                        tentative = 1
                if tentative == 1:
                    heuri = self.HeuristicCost(next)
                    f = new_cost + heuri
                    tmp = codaE(next, f)
                    open.insert(tmp)
                    came_from[next.getNome()] = current.getG().getNome()
            close.append(current)
        self.reconstruct_path(came_from, self.goal.getNome())

    
    def reconstruct_path(self,came_from,current_node):
        if current_node != "Entrata":
            if current_node == "Uscita":
                print(current_node)
                tmp = came_from[current_node]
                print(tmp)
                if current_node != "Entrata":
                    self.reconstruct_path(came_from,came_from[current_node]) 
            else:
                tmp = came_from[current_node]
                print(tmp)
                if tmp != "Entrata":
                    self.reconstruct_path(came_from,came_from[current_node])


#tmp = Astar(mentrata, muscita)
#es = tmp.svolgimento()
```
